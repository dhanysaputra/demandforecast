This notebook explores raw Purchase Order and Sales Order data.

Goals:
- Load historical data (CSV or Salesforce extract)
- Clean timestamps
- Check missing data
- Visualize basic trends
- Understand demand intermittency and variability


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

purchase_df = pd.read_pickle("../data/purchase_df.pkl")
sales_df = pd.read_pickle("../data/sales_df.pkl")

purchase_df.head(), sales_df.head()
```

## Plot Purchase vs Sales

```python
purchase_ts = purchase_df.groupby(pd.Grouper(key="DeliveryDate", freq="M"))['DeliveredQuantity'].sum()
sales_ts = sales_df.groupby(pd.Grouper(key="DeliveryDate", freq="M"))['DeliveredQuantity'].sum()

plt.figure(figsize=(14,6))
plt.plot(purchase_ts, label='Purchased Qty')
plt.plot(sales_ts, label='Sales Qty')
plt.legend()
plt.title('Raw Purchase vs Sales Over Time')
plt.show()
```
